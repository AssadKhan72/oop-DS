```python
import math

class Point:

    def __init__(self, x: float = 0, y: float = 0) -> None:
        self._x = x
        self._y = y

    @property
    def x(self) -> float:
        return self._x

    @x.setter
    def x(self, value: float) -> None:
        self._x = value

    @property
    def y(self) -> float:
        return self._y

    @y.setter
    def y(self, value: float) -> None:
        self._y = value

    def distance(self, other: 'Point') -> float:
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def distance_from_origin(self) -> float:
        return math.sqrt(self.x**2 + self.y**2)

    def locate(self) -> str:
        if self.x == 0 and self.y == 0:
            return "Origin"
        elif self.x == 0:
            return "y-axis"
        elif self.y == 0:
            return "x-axis"
        elif self.x > 0 and self.y > 0:
            return "Quadrant I"
        elif self.x < 0 and self.y > 0:
            return "Quadrant II"
        elif self.x < 0 and self.y < 0:
            return "Quadrant III"
        else:  # self.x > 0 and self.y < 0
            return "Quadrant IV"

    def __repr__(self) -> str:
        return f"Point(x={self.x}, y={self.y})"

    def __str__(self) -> str:
        return f"({self.x}, {self.y})"

    @staticmethod
    def distance_between_two_points(point1: 'Point', point2: 'Point') -> float:

        return math.sqrt((point1.x - point2.x)**2 + (point1.y - point2.y)**2)

    @classmethod
    def from_tuple(cls, coordinates: tuple[float, float]) -> 'Point':
 
        if not isinstance(coordinates, tuple):
            raise TypeError("Coordinates must be a tuple")
        if len(coordinates) != 2:
            raise ValueError("Coordinates tuple must contain two values.")
        x, y = coordinates
        return cls(x, y)

    @property
    def magnitude(self) -> float:
        """Calculates the magnitude (distance from origin) of the point."""
        return self.distance_from_origin()

```

```python
# Example Usage
p1 = Point(3, 4)
p2 = Point(0, 0)

print(p1)
print(p2)

print(f"Distance between p1 and p2: {p1.distance(p2)}")
print(f"Distance of p1 from origin: {p1.distance_from_origin()}")
print(f"Location of p1: {p1.locate()}")

# Static method example
dist_static = Point.distance_between_two_points(p1, p2)
print(f"Distance (using static method): {dist_static}")

# Class method example
p3 = Point.from_tuple((5, 12))
print(f"Point created from tuple: {p3}")
print(f"p3's x-coordinate: {p3.x}")
print(f"p3's y-coordinate: {p3.y}")

#Magnitude property example
print(f"Magnitude of p1: {p1.magnitude}")

#Using Properties
p1.x = 10
print(p1)
p1.y = 20
print(p1)

```

```python
import math

class Circle:

    def __init__(self, radius: float = 1) -> None:
        if radius <= 0:
            raise ValueError("Radius must be positive.")
        self._radius = radius

    @property
    def radius(self) -> float:
        return self._radius

    @radius.setter
    def radius(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Radius must be positive.")
        self._radius = value

    def cylinder_volume(self, height: float) -> float:
        if height <= 0:
            raise ValueError("Height must be positive.")
        return self.area * height

    @property
    def area(self) -> float:
        return math.pi * self._radius**2

    @property
    def circumference(self) -> float:
        return 2 * math.pi * self._radius

    @property
    def diameter(self) -> float:
        return 2 * self._radius

    def __repr__(self) -> str:
        return f"Circle(radius={self._radius})"

    def __str__(self) -> str:
        return f"Circle with radius: {self._radius}"

    @staticmethod
    def calculate_area_from_radius(radius: float) -> float:
        if radius <= 0:
            raise ValueError("Radius must be positive.")
        return math.pi * radius**2

    @classmethod
    def from_diameter(cls, diameter: float) -> 'Circle':
        if diameter <= 0:
            raise ValueError("Diameter must be positive.")
        radius = diameter / 2
        return cls(radius)




```

```python
circle1 = Circle(5)
print(circle1)
print(f"Area: {circle1.area}")
print(f"Circumference: {circle1.circumference}")
print(f"Diameter: {circle1.diameter}")

# Using the static method
area_from_radius = Circle.calculate_area_from_radius(7)
print(f"Area calculated using static method (radius=7): {area_from_radius}")

# Using the class method
circle2 = Circle.from_diameter(10)
print(circle2)
print(f"Radius from diameter: {circle2.radius}")

#Using Properties
circle1.radius = 8
print(circle1)
print(f"Area: {circle1.area}")
```

```python
import cmath
import math

class RLC:

    def __init__(self, resistance: float = 0, inductance: float = 0, capacitance: float = 0, angular_frequency: float = 1) -> None:
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if inductance < 0:
            raise ValueError("Inductance must be non-negative.")
        if capacitance < 0:
            raise ValueError("Capacitance must be non-negative.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        """Gets the resistance."""
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        """Sets the resistance."""
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        """Gets the inductance."""
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        """Sets the inductance."""
        if value < 0:
            raise ValueError("Inductance must be non-negative.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        """Gets the capacitance."""
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        """Sets the capacitance."""
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value
    
    @property
    def angular_frequency(self) -> float:
        """Gets the angular frequency."""
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        """Sets the angular frequency."""
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def impedance(self) -> complex:
        j = complex(0, 1)
        omega = self._angular_frequency
        if self._inductance == 0 and self._capacitance == 0:
            return complex(self._resistance)
        return self._resistance + j * (omega * self._inductance - 1 / (omega * self._capacitance))

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    @property
    def power_factor(self) -> float:
        return math.cos(self.phase)

    def current(self, voltage: complex) -> complex:
        if self.impedance == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance

    def __repr__(self) -> str:
        """Returns a string representation of the RLC object for debugging."""
        return f"RLC(R={self._resistance}, L={self._inductance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the RLC object."""
        return f"RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

    @staticmethod
    def calculate_reactance(angular_frequency: float, inductance: float, capacitance: float) -> complex:
      if angular_frequency <= 0:
          raise ValueError("Angular frequency must be positive.")
      j = complex(0,1)
      if inductance == 0 and capacitance == 0:
          return complex(0)
      return j * (angular_frequency * inductance - 1 / (angular_frequency * capacitance))

    @classmethod
    def from_component_values(cls, components: dict, angular_frequency: float = 1) -> 'RLC':
      try:
          resistance = components['R']
          inductance = components['L']
          capacitance = components['C']
      except KeyError as e:
          raise KeyError(f"Missing component value: {e}")
      return cls(resistance, inductance, capacitance, angular_frequency)

```

```python
class Square:


    def __init__(self, side: float = 1) -> None:
        if side <= 0:
            raise ValueError("Side length must be positive.")
        self._side = side

    @property
    def side(self) -> float:
        """Gets the side length."""
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        self._side = value

    @property
    def area(self) -> float:
        """Calculates the area of the square."""
        return self._side**2

    @property
    def perimeter(self) -> float:
        """Calculates the perimeter of the square."""
        return 4 * self._side

    @property
    def diagonal(self) -> float:
        """Calculates the length of the diagonal."""
        return self._side * 2**0.5  # or self._side * math.sqrt(2)

    def __repr__(self) -> str:
        """Returns a string representation of the Square object for debugging."""
        return f"Square(side={self._side})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the Square object."""
        return f"Square with side length: {self._side}"

    @staticmethod
    def area_from_side(side: float) -> float:
        if side <= 0:
            raise ValueError("Side length must be positive.")
        return side**2

    @classmethod
    def from_area(cls, area: float) -> 'Square':
        if area <= 0:
            raise ValueError("Area must be positive.")
        side = area**0.5  # or math.sqrt(area)
        return cls(side)



```

```python
square1 = Square(5)
print(square1)
print(f"Area: {square1.area}")
print(f"Perimeter: {square1.perimeter}")
print(f"Diagonal: {square1.diagonal}")

#Demonstrating static method
area_static = Square.area_from_side(6)
print(f"Area using static method: {area_static}")

#Demonstrating class method
square2 = Square.from_area(25)
print(square2)
print(f"Side from area: {square2.side}")

#Using Properties
square1.side = 10
print(square1)
print(f"Area: {square1.area}")

#Demonstrating default values
square3 = Square()
print(square3)
```

```python
class Rectangle:

    def __init__(self, length: float = 1, width: float = 1) -> None:
        if length <= 0:
            raise ValueError("Length must be positive.")
        if width <= 0:
            raise ValueError("Width must be positive.")
        self._length = length
        self._width = width

    @property
    def length(self) -> float:
        """Gets the length of the rectangle."""
        return self._length

    @length.setter
    def length(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Length must be positive.")
        self._length = value

    @property
    def width(self) -> float:
        """Gets the width of the rectangle."""
        return self._width

    @width.setter
    def width(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Width must be positive.")
        self._width = value

    @property
    def area(self) -> float:
        """Calculates the area of the rectangle."""
        return self._length * self._width

    @property
    def perimeter(self) -> float:
        """Calculates the perimeter of the rectangle."""
        return 2 * (self._length + self._width)

    @property
    def diagonal(self) -> float:
        """Calculates the length of the diagonal."""
        return (self._length**2 + self._width**2)**0.5 # or math.sqrt(self._length**2 + self._width**2)

    def __repr__(self) -> str:
        """Returns a string representation of the Rectangle object for debugging."""
        return f"Rectangle(length={self._length}, width={self._width})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the Rectangle object."""
        return f"Rectangle with length: {self._length}, width: {self._width}"

    @staticmethod
    def area_from_length_width(length: float, width: float) -> float:
        if length <= 0:
            raise ValueError("Length must be positive.")
        if width <= 0:
            raise ValueError("Width must be positive.")
        return length * width

    @classmethod
    def from_area_and_length(cls, area: float, length: float) -> 'Rectangle':
        if area <= 0:
            raise ValueError("Area must be positive.")
        if length <= 0:
            raise ValueError("Length must be positive.")
        if length > area:
            raise ValueError("Length cannot be greater than area")
        width = area / length
        return cls(length, width)


```

```python
rect1 = Rectangle(5, 10)
print(rect1)
print(f"Area: {rect1.area}")
print(f"Perimeter: {rect1.perimeter}")
print(f"Diagonal: {rect1.diagonal}")

#Demonstrating static method
area_static = Rectangle.area_from_length_width(6, 4)
print(f"Area using static method: {area_static}")

#Demonstrating class method
rect2 = Rectangle.from_area_and_length(50, 5)
print(rect2)
print(f"Width from area and length: {rect2.width}")

#Using Properties
rect1.length = 10
print(rect1)
print(f"Area: {rect1.area}")

#Demonstrating default values
rect3 = Rectangle()
print(rect3)
```

```python
import math

class Triangle:


    def __init__(self, a: float, b: float, c: float) -> None:
        if not all(side > 0 for side in (a, b, c)):
            raise ValueError("Side lengths must be positive.")
        if not (a + b > c and a + c > b and b + c > a):
            raise ValueError("Triangle inequality theorem not satisfied.")
        self._a = a
        self._b = b
        self._c = c

    @property
    def a(self) -> float:
        """Gets the length of side a."""
        return self._a

    @a.setter
    def a(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (value + self._b > self._c and value + self._c > self._b and self._b + self._c > value):
            raise ValueError("Triangle inequality theorem not satisfied.")
        self._a = value

    @property
    def b(self) -> float:
        """Gets the length of side b."""
        return self._b

    @b.setter
    def b(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (self._a + value > self._c and self._a + self._c > value and value + self._c > self._a):
            raise ValueError("Triangle inequality theorem not satisfied.")
        self._b = value

    @property
    def c(self) -> float:
        """Gets the length of side c."""
        return self._c

    @c.setter
    def c(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (self._a + self._b > value and self._a + value > self._b and self._b + value > self._a):
            raise ValueError("Triangle inequality theorem not satisfied.")
        self._c = value

    @property
    def area(self) -> float:
        """Calculates the area of the triangle using Heron's formula."""
        s = (self._a + self._b + self._c) / 2
        return math.sqrt(s * (s - self._a) * (s - self._b) * (s - self._c))

    @property
    def perimeter(self) -> float:
        """Calculates the perimeter of the triangle."""
        return self._a + self._b + self._c

    def __repr__(self) -> str:
        """Returns a string representation of the Triangle object for debugging."""
        return f"Triangle(a={self._a}, b={self._b}, c={self._c})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the Triangle object."""
        return f"Triangle with sides: a={self._a}, b={self._b}, c={self._c}"

    @staticmethod
    def is_valid_triangle(a: float, b: float, c: float) -> bool:
        if not all(side > 0 for side in (a, b, c)):
            return False
        return a + b > c and a + c > b and b + c > a

    @classmethod
    def from_area_and_base_height(cls, area: float, base: float, height: float) -> 'Triangle':
        if area <= 0:
            raise ValueError("Area must be positive.")
        if base <= 0:
            raise ValueError("Base must be positive.")
        if height <= 0:
            raise ValueError("Height must be positive.")

        #Calculate the other sides using pythagoras theorem given base and height
        side1 = math.sqrt((base/2)**2 + height**2)
        return cls(base, side1, side1) #This will create an isoceles triangle



```

```python
# Example Usage
triangle1 = Triangle(3, 4, 5)  # A right-angled triangle
print(triangle1)
print(f"Area: {triangle1.area}")
print(f"Perimeter: {triangle1.perimeter}")

#Demonstrating static method
is_valid = Triangle.is_valid_triangle(3, 4, 5)
print(f"Is (3, 4, 5) a valid triangle? {is_valid}")

is_valid = Triangle.is_valid_triangle(1, 2, 5)
print(f"Is (1, 2, 5) a valid triangle? {is_valid}")

#Demonstrating class method
triangle2 = Triangle.from_area_and_base_height(10, 5, 4)
print(triangle2)

#Using Properties
triangle1.a = 6
print(triangle1)
```

```python
class Trapezoid:

    def __init__(self, base1: float, base2: float, height: float, side1: float = 0, side2: float = 0) -> None:
        if not all(x > 0 for x in (base1, base2, height)):
            raise ValueError("Base and height must be positive.")
        if base1 == base2:
            raise ValueError("Bases cannot be equal. Use Square or Rectangle class if needed.")
        if side1 < 0 or side2 < 0:
            raise ValueError("Sides must be non-negative.")

        self._base1 = base1
        self._base2 = base2
        self._height = height
        self._side1 = side1
        self._side2 = side2

    @property
    def base1(self) -> float:
        """Gets the length of the first base."""
        return self._base1

    @base1.setter
    def base1(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Base must be positive.")
        self._base1 = value

    @property
    def base2(self) -> float:
        return self._base2

    @base2.setter
    def base2(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Base must be positive.")
        self._base2 = value

    @property
    def height(self) -> float:
        return self._height

    @height.setter
    def height(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Height must be positive.")
        self._height = value

    @property
    def side1(self) -> float:
        return self._side1

    @side1.setter
    def side1(self, value: float) -> None:
        if value < 0:
            raise ValueError("Side must be non-negative.")
        self._side1 = value

    @property
    def side2(self) -> float:
        return self._side2

    @side2.setter
    def side2(self, value: float) -> None:
        if value < 0:
            raise ValueError("Side must be non-negative.")
        self._side2 = value

    @property
    def area(self) -> float:
        return 0.5 * (self._base1 + self._base2) * self._height

    def perimeter(self) -> float:
        return self._base1 + self._base2 + self._side1 + self._side2

    def __repr__(self) -> str:
        return f"Trapezoid(base1={self._base1}, base2={self._base2}, height={self._height}, side1={self._side1}, side2={self._side2})"

    def __str__(self) -> str:
        return f"Trapezoid with bases: {self._base1}, {self._base2}, height: {self._height}, sides: {self._side1}, {self._side2}"

    @staticmethod
    def area_from_bases_height(base1: float, base2: float, height: float) -> float:
        if not all(x > 0 for x in (base1, base2, height)):
            raise ValueError("Base and height must be positive.")
        return 0.5 * (base1 + base2) * height

    @classmethod
    def from_area_bases(cls, area: float, base1: float, base2: float) -> 'Trapezoid':
        if not all(x > 0 for x in (area, base1, base2)):
            raise ValueError("Area and bases must be positive.")
        if base1 == base2:
            raise ValueError("Bases cannot be equal. Use Square or Rectangle class if needed.")

        height = (2 * area) / (base1 + base2)
        return cls(base1, base2, height)



```

```python
class Parallelogram:


    def __init__(self, base: float, side: float, height: float) -> None:
        if not all(x > 0 for x in (base, side, height)):
            raise ValueError("Base, side, and height must be positive.")
        self._base = base
        self._side = side
        self._height = height

    @property
    def base(self) -> float:
        return self._base

    @base.setter
    def base(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Base must be positive.")
        self._base = value

    @property
    def side(self) -> float:
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side must be positive.")
        self._side = value

    @property
    def height(self) -> float:
        return self._height

    @height.setter
    def height(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Height must be positive.")
        self._height = value

    @property
    def area(self) -> float:
        return self._base * self._height

    @property
    def perimeter(self) -> float:
        return 2 * (self._base + self._side)

    def __repr__(self) -> str:
        return f"Parallelogram(base={self._base}, side={self._side}, height={self._height})"

    def __str__(self) -> str:
        return f"Parallelogram with base: {self._base}, side: {self._side}, height: {self._height}"

    @staticmethod
    def area_from_base_height(base: float, height: float) -> float:
        if not all(x > 0 for x in (base, height)):
            raise ValueError("Base and height must be positive.")
        return base * height

    @classmethod
    def from_area_base(cls, area: float, base: float) -> 'Parallelogram':
        if not all(x > 0 for x in (area, base)):
            raise ValueError("Area and base must be positive.")
        height = area / base
        return cls(base, base, height) #This will create a rectangle


```

```python
import math

class Rhombus:

    def __init__(self, side: float, diagonal1: float = 0, diagonal2: float = 0) -> None:
        if side <= 0:
            raise ValueError("Side length must be positive.")
        if diagonal1 < 0 or diagonal2 < 0:
            raise ValueError("Diagonals must be non-negative.")
        if diagonal1 > 0 and diagonal2 > 0: #Check if the diagonals are valid if they are provided
          if not (diagonal1**2 + diagonal2**2 == 4*side**2):
            raise ValueError("Diagonals do not satisfy the rhombus diagonal properties.")
        self._side = side
        self._diagonal1 = diagonal1
        self._diagonal2 = diagonal2

    @property
    def side(self) -> float:
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if self._diagonal1 > 0 and self._diagonal2 > 0: #Check if the diagonals are valid if they are provided
          if not (self._diagonal1**2 + self._diagonal2**2 == 4*value**2):
            raise ValueError("Diagonals do not satisfy the rhombus diagonal properties.")
        self._side = value

    @property
    def diagonal1(self) -> float:
        return self._diagonal1

    @diagonal1.setter
    def diagonal1(self, value: float) -> None:
        if value < 0:
            raise ValueError("Diagonal must be non-negative.")
        if self._side > 0 and self._diagonal2 > 0: #Check if the diagonals are valid if they are provided
          if not (value**2 + self._diagonal2**2 == 4*self._side**2):
            raise ValueError("Diagonals do not satisfy the rhombus diagonal properties.")
        self._diagonal1 = value

    @property
    def diagonal2(self) -> float:
        return self._diagonal2

    @diagonal2.setter
    def diagonal2(self, value: float) -> None:
        if value < 0:
            raise ValueError("Diagonal must be non-negative.")
        if self._side > 0 and self._diagonal1 > 0: #Check if the diagonals are valid if they are provided
          if not (self._diagonal1**2 + value**2 == 4*self._side**2):
            raise ValueError("Diagonals do not satisfy the rhombus diagonal properties.")
        self._diagonal2 = value

    @property
    def area(self) -> float:
        if self._diagonal1 == 0 or self._diagonal2 == 0:
          return 0
        return 0.5 * self._diagonal1 * self._diagonal2

    @property
    def perimeter(self) -> float:
        return 4 * self._side

    def __repr__(self) -> str:
        return f"Rhombus(side={self._side}, diagonal1={self._diagonal1}, diagonal2={self._diagonal2})"

    def __str__(self) -> str:
        return f"Rhombus with side: {self._side}, diagonal1: {self._diagonal1}, diagonal2: {self._diagonal2}"

    @staticmethod
    def area_from_diagonals(diagonal1: float, diagonal2: float) -> float:
        if diagonal1 < 0 or diagonal2 < 0:
            raise ValueError("Diagonals must be non-negative.")
        return 0.5 * diagonal1 * diagonal2

    @classmethod
    def from_side_diagonal(cls, side: float, diagonal1: float) -> 'Rhombus':
        if side <= 0:
            raise ValueError("Side length must be positive.")
        if diagonal1 < 0:
            raise ValueError("Diagonal must be non-negative.")
        if diagonal1 > 2*side:
            raise ValueError("Diagonal cannot be greater than 2 times the side")
        diagonal2 = math.sqrt(4 * side**2 - diagonal1**2)
        return cls(side, diagonal1, diagonal2)



```

```python
import cmath

class Resistor:

    def __init__(self, resistance: float = 0) -> None:
        if resistance < 0:
            raise ValueError("Resistance cannot be negative.")
        self._resistance = resistance

    @property
    def resistance(self) -> float:
        """Gets the resistance value."""
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        """Sets the resistance value."""
        if value < 0:
            raise ValueError("Resistance cannot be negative.")
        self._resistance = value

    @property
    def impedance(self) -> complex:
        return complex(self._resistance)  # Impedance of a resistor is purely real

    def voltage(self, current: complex) -> complex:
        return current * self.impedance

    def current(self, voltage: complex) -> complex:
        if self.resistance == 0:
            raise ZeroDivisionError("Resistance cannot be zero for current calculation.")
        return voltage / self.impedance

    def power(self, current: complex) -> float:
        return abs(current)**2 * self.resistance #abs() is used to get the magnitude of the complex number

    def __repr__(self) -> str:
        """Returns a string representation of the Resistor object for debugging."""
        return f"Resistor(resistance={self._resistance})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the Resistor object."""
        return f"Resistor: {self._resistance} Ω"

    @staticmethod
    def power_from_voltage_resistance(voltage: complex, resistance: float) -> float:
        """Calculates power from voltage and resistance (static method)."""
        if resistance <= 0:
            raise ValueError("Resistance must be positive.")
        return abs(voltage)**2 / resistance



```

```python
import cmath
import math

class RC:

    def __init__(self, resistance: float = 0, capacitance: float = 0, angular_frequency: float = 1) -> None:
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if capacitance < 0:
            raise ValueError("Capacitance must be non-negative.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._capacitance = capacitance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        """Gets the resistance."""
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        """Sets the resistance."""
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def capacitance(self) -> float:
        """Gets the capacitance."""
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        """Sets the capacitance."""
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value

    @property
    def angular_frequency(self) -> float:
        """Gets the angular frequency."""
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        """Sets the angular frequency."""
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def impedance(self) -> complex:
        j = complex(0, 1)
        if self._capacitance == 0:
          return complex(self._resistance)
        return self._resistance - j / (self._angular_frequency * self._capacitance)

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    def voltage(self, current: complex) -> complex:
        return current * self.impedance

    def current(self, voltage: complex) -> complex:
        if self.impedance == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance

    def power_dissipated(self, current: complex) -> float:
        return (abs(current)**2) * self._resistance

    def __repr__(self) -> str:
        """Returns a string representation of the RC object for debugging."""
        return f"RC(R={self._resistance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the RC object."""
        return f"RC Circuit: R={self._resistance} Ω, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

    @staticmethod
    def calculate_capacitive_reactance(capacitance: float, angular_frequency: float) -> float:
      if angular_frequency <= 0:
          raise ValueError("Angular frequency must be positive.")
      if capacitance < 0:
          raise ValueError("Capacitance must be non-negative.")
      if capacitance == 0:
          return float('inf')
      return -1 / (angular_frequency * capacitance)

    @classmethod
    def from_impedance_resistance(cls, impedance: complex, resistance: float, angular_frequency: float) -> 'RC':
      if angular_frequency <= 0:
          raise ValueError("Angular frequency must be positive.")
      if resistance < 0:
          raise ValueError("Resistance must be non-negative.")
      if not math.isclose(impedance.real, resistance):
          raise ValueError("The real part of the impedance does not match the provided resistance.")
      capacitance = -1 / (angular_frequency * impedance.imag)
      return cls(resistance, capacitance, angular_frequency)


```

```python
import cmath
import math

class RLCParallel:
    """Represents a parallel RLC circuit."""

    def __init__(self, resistance: float = 0, inductance: float = 0, capacitance: float = 0, angular_frequency: float = 1) -> None:
        """Initializes a parallel RLC circuit.

        Args:
            resistance (float, optional): Resistance (in ohms Ω). Defaults to 0.
            inductance (float, optional): Inductance (in henries H). Defaults to 0.
            capacitance (float, optional): Capacitance (in farads F). Defaults to 0.
            angular_frequency (float, optional): Angular frequency (in radians per second rad/s). Defaults to 1.

        Raises:
            ValueError: If resistance, inductance, or capacitance is negative, or if angular frequency is not positive.
        """
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if inductance < 0:
            raise ValueError("Inductance must be non-negative.")
        if capacitance < 0:
            raise ValueError("Capacitance must be non-negative.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        """Gets the resistance."""
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        """Sets the resistance."""
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        """Gets the inductance."""
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        """Sets the inductance."""
        if value < 0:
            raise ValueError("Inductance must be non-negative.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        """Gets the capacitance."""
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        """Sets the capacitance."""
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value

    @property
    def angular_frequency(self) -> float:
        """Gets the angular frequency."""
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        """Sets the angular frequency."""
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def admittance(self) -> complex:
        """Calculates the admittance (Y) of the parallel RLC circuit.

        Returns:
            complex: The admittance (1/R + j(ωC - 1/(ωL))).
        """
        j = complex(0, 1)
        omega = self._angular_frequency
        if self._resistance == 0:
            conductance = 0
        else:
            conductance = 1 / self._resistance
        if self._inductance == 0:
            susceptance_l = 0
        else:
            susceptance_l = -1 / (omega * self._inductance)
        if self._capacitance == 0:
            susceptance_c = 0
        else:
            susceptance_c = omega * self._capacitance
        return conductance + j * (susceptance_c + susceptance_l)

    @property
    def impedance(self) -> complex:
        """Calculates the impedance (Z) of the parallel RLC circuit.

        Returns:
            complex: The impedance (1/Y).
        Raises:
            ZeroDivisionError: If the admittance is zero.
        """
        if self.admittance == 0:
            raise ZeroDivisionError("Admittance cannot be zero.")
        return 1 / self.admittance

    @property
    def phase(self) -> float:
        """Calculates the phase angle (theta) of the impedance in radians.

        Returns:
            float: The phase angle in radians.
        """
        return cmath.phase(self.impedance)

    def voltage(self, current: complex) -> complex:
        """Calculates the voltage across the parallel RLC circuit (V = I * Z).

        Args:
            current (complex): The current flowing into the circuit.

        Returns:
            complex: The voltage across the circuit.
        """
        return current * self.impedance

    def current(self, voltage: complex) -> complex:
        """Calculates the current flowing into the parallel RLC circuit (I = V * Y).

        Args:
            voltage (complex): The voltage across the circuit.

        Returns:
            complex: The current flowing into the circuit.
        """
        return voltage * self.admittance

    def power_dissipated(self, voltage: complex) -> float:
        """Calculates the power dissipated by the resistor (P = V^2 / R).

        Args:
            voltage (complex): The voltage across the circuit.

        Returns:
            float: The power dissipated by the resistor.
        """
        if self._resistance == 0:
          return 0
        return (abs(voltage)**2) / self._resistance

    def __repr__(self) -> str:
        """Returns a string representation of the parallel RLC object for debugging."""
        return f"RLCParallel(R={self._resistance}, L={self._inductance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the parallel RLC object."""
        return f"Parallel RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

    @staticmethod
    def calculate_susceptance(inductance: float, capacitance: float, angular_frequency: float) -> complex:
  
      if angular_frequency <= 0:
          raise ValueError("Angular frequency must be positive.")
      if inductance < 0:
          raise ValueError("Inductance must be non-negative.")
      if capacitance < 0:
          raise ValueError("Capacitance must be non-negative.")
      j = complex(0,1)
      if inductance == 0:
        susceptance_l = 0
      else:
        susceptance_l = -1/(angular_frequency * inductance)
      if capacitance == 0:
        susceptance_c = 0
      else:
        susceptance_c = angular_frequency * capacitance
      return j * (susceptance_c + susceptance_l)

    @classmethod
    def from_impedance_resistance(cls, impedance: complex, resistance: float, angular_frequency: float) -> 'RLCParallel':
 
      if angular_frequency <= 0:
          raise ValueError("Angular frequency must be positive.")
      if resistance < 0:
          raise ValueError("Resistance must be non-negative.")
      conductance = 1/resistance
      if not math.isclose((1/impedance).real, conductance):
          raise ValueError("The real part of the inverse of the impedance does not match the provided conductance.")
      susceptance = (1/impedance).imag
      capacitance = (susceptance / 2 + math.sqrt((susceptance/2)**2 + (1/(angular_frequency**2 * 0.000000000000001)))) / angular_
```

```python
import cmath
import math

class RLCSeries:


    def __init__(self, resistance: float = 0, inductance: float = 0, capacitance: float = 0) -> None:
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if inductance < 0:
            raise ValueError("Inductance must be non-negative.")
        if capacitance < 0:
            raise ValueError("Capacitance must be non-negative.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance

    @property
    def resistance(self) -> float:
        """Gets the resistance."""
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        """Sets the resistance."""
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        """Gets the inductance."""
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        """Sets the inductance."""
        if value < 0:
            raise ValueError("Inductance must be non-negative.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        """Gets the capacitance."""
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        """Sets the capacitance."""
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value

    def impedance(self, angular_frequency: float) -> complex:
        j = complex(0, 1)
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        if self._inductance == 0 and self._capacitance == 0:
            return complex(self._resistance)
        return self._resistance + j * (angular_frequency * self._inductance - 1 / (angular_frequency * self._capacitance))

    def phase(self, angular_frequency: float) -> float:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        return cmath.phase(self.impedance(angular_frequency))

    def current(self, voltage: complex, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        if self.impedance(angular_frequency) == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance(angular_frequency)

    def resonant_frequency(self) -> float:
        if self._inductance == 0 or self._capacitance == 0:
            raise ValueError("Inductance and Capacitance cannot be zero to calculate resonant frequency.")
        return 1 / math.sqrt(self._inductance * self._capacitance)

    def impedance_at_resonance(self) -> complex:
        return self.impedance(self.resonant_frequency())

    def __repr__(self) -> str:
        """Returns a string representation of the RLC object for debugging."""
        return f"RLCSeries(R={self._resistance}, L={self._inductance}, C={self._capacitance})"

    def __str__(self) -> str:
        """Returns a user-friendly string representation of the RLC object."""
        return f"Series RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F"

```
