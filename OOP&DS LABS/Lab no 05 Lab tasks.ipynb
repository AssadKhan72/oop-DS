```python
import math

class Point:


    def __init__(self, x: float = 0, y: float = 0) -> None:
        self._x = x  # Use protected attributes
        self._y = y

    @property
    def x(self) -> float:
        return self._x

    @x.setter
    def x(self, value: float) -> None:
        self._x = value

    @property
    def y(self) -> float:
        return self._y

    @y.setter
    def y(self, value: float) -> None:
        self._y = value

    def distance(self, other: 'Point') -> float:
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def distance_from_origin(self) -> float:
        return math.sqrt(self.x**2 + self.y**2)

    def locate(self) -> str:
        if self.x == 0 and self.y == 0:
            return "Origin"
        elif self.x == 0:
            return "y-axis"
        elif self.y == 0:
            return "x-axis"
        elif self.x > 0 and self.y > 0:
            return "Quadrant I"
        elif self.x < 0 and self.y > 0:
            return "Quadrant II"
        elif self.x < 0 and self.y < 0:
            return "Quadrant III"
        else:  # self.x > 0 and self.y < 0
            return "Quadrant IV"

    def display(self) -> None:
      
        print(f"Point coordinates: (x = {self.x}, y = {self.y})")

    def __repr__(self) -> str:
        return f"Point(x={self.x}, y={self.y})"

    def __str__(self) -> str:
        return f"({self.x}, {self.y})"


# a. Define inst_1 and pass two numbers.
inst_1 = Point(3.0, 4.0)

# b. Make another instance inst_2.
inst_2 = Point()

# c. Print inst_1 and inst_2.
print(inst_1)
print(inst_2)

# d. Also print the location of both points.
inst_1.display()
print(f"Location: {inst_1.locate()}")
inst_2.display()
print(f"Location: {inst_2.locate()}")

# e. Calculate the distance between these two points.
distance = inst_1.distance(inst_2)
print(f"Distance between inst_1 and inst_2: {distance}")

# f. Call the __dict__ by the class name.
print(Point.__dict__)

# g. Also pass the class name to the vars built-in function.
print(vars(Point))

# h. Call the __dict__ on the object of the class.
print(inst_1.__dict__)

# i. Pass the class name to help.
help(Point)



# k. Modify the __init__ by making its parameters default and verify by instances.
inst_3 = Point()     
inst_4 = Point(5)    
inst_5 = Point(y=10) 
print(inst_3)
print(inst_4)
print(inst_5)

#Demonstrating the properties
inst_1.x = 10
print(inst_1)
inst_1.y = 20
print(inst_1)
```

```python
import math

class Circle:

    def __init__(self, radius: float = 1) -> None:
        if radius <= 0:
            raise ValueError("Radius must be positive.")
        self._radius = radius  # Use a protected attribute for radius

    @property
    def radius(self) -> float:
        return self._radius

    @radius.setter
    def radius(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Radius must be positive.")
        self._radius = value

    def cylinder_volume(self, height: float) -> float:
        if height <= 0:
            raise ValueError("Height must be positive.")
        return self.area * height

    @property
    def area(self) -> float:
        return math.pi * self._radius**2

    @property
    def circumference(self) -> float:
        return 2 * math.pi * self._radius

    @property
    def diameter(self) -> float:
        return 2 * self._radius

    def __repr__(self) -> str:
        return f"Circle(radius={self._radius})"

    def __str__(self) -> str:
        return f"Circle with radius: {self._radius}"


```

```python
import cmath
import math

class RLC:


    def __init__(self, resistance: float = 0, inductance: float = 0, capacitance: float = 0) -> None:
        if resistance < 0 or inductance < 0 or capacitance < 0:
            raise ValueError("Resistance, inductance, and capacitance must be non-negative.")
        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance
        self._angular_frequency = 1

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Inductance must be non-negative.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value
    
    @property
    def angular_frequency(self) -> float:
        return self._angular_frequency
    
    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def impedance(self) -> complex:
        j = complex(0, 1)  # Imaginary unit
        omega = self.angular_frequency
        if self.inductance == 0 and self.capacitance == 0:
          return complex(self.resistance)
        return self.resistance + j * (omega * self.inductance - 1 / (omega * self.capacitance))

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    @property
    def power_factor(self) -> float:
        return math.cos(self.phase)

    def current(self, voltage: complex) -> complex:
        if self.impedance == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance

    def __repr__(self) -> str:
        return f"RLC(R={self._resistance}, L={self._inductance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        return f"RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

```

```python
class Square:
  

    def __init__(self, side: float) -> None:
        if side <= 0:
            raise ValueError("Side length must be positive.")
        self._side = side

    @property
    def side(self) -> float:
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        self._side = value

    @property
    def area(self) -> float:
        return self._side**2

    @property
    def perimeter(self) -> float:
        return 4 * self._side

    @property
    def diagonal(self) -> float:
        return self._side * math.sqrt(2)

    def __repr__(self) -> str:
        return f"Square(side={self._side})"

    def __str__(self) -> str:
        return f"Square with side length: {self._side}"


# Example usage:
square1 = Square(5)
print(square1)
print(f"Area: {square1.area}")
print(f"Perimeter: {square1.perimeter}")
print(f"Diagonal: {square1.diagonal}")

square2 = Square(10)
print(square2)

#Using Properties
square1.side = 7
print(square1)
print(f"Area: {square1.area}")

```

```python
class Rectangle:

    def __init__(self, length: float, width: float) -> None:
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive.")
        self._length = length
        self._width = width

    @property
    def length(self) -> float:
        return self._length

    @length.setter
    def length(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Length must be positive.")
        self._length = value

    @property
    def width(self) -> float:
        return self._width

    @width.setter
    def width(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Width must be positive.")
        self._width = value

    @property
    def area(self) -> float:
        return self._length * self._width

    @property
    def perimeter(self) -> float:
        return 2 * (self._length + self._width)

    @property
    def diagonal(self) -> float:
        return math.sqrt(self._length**2 + self._width**2)

    def __repr__(self) -> str:
        return f"Rectangle(length={self._length}, width={self._width})"

    def __str__(self) -> str:
        return f"Rectangle with length: {self._length}, width: {self._width}"


# Example Usage
rect1 = Rectangle(5, 10)
print(rect1)
print(f"Area: {rect1.area}")
print(f"Perimeter: {rect1.perimeter}")
print(f"Diagonal: {rect1.diagonal}")

rect2 = Rectangle(7, 3)
print(rect2)

#Using Properties
rect1.length = 12
print(rect1)
print(f"Area: {rect1.area}")
rect1.width = 6
print(rect1)
print(f"Perimeter: {rect1.perimeter}")
```

```python
import math

class Triangle:

    def __init__(self, side1: float, side2: float, side3: float) -> None:
        if not (side1 > 0 and side2 > 0 and side3 > 0):
            raise ValueError("Side lengths must be positive.")
        if not (side1 + side2 > side3 and side1 + side3 > side2 and side2 + side3 > side1):
            raise ValueError("Invalid triangle: Side lengths do not satisfy the triangle inequality theorem.")

        self._side1 = side1
        self._side2 = side2
        self._side3 = side3

    @property
    def side1(self) -> float:
        return self._side1

    @side1.setter
    def side1(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (value + self._side2 > self._side3 and value + self._side3 > self._side2 and self._side2 + self._side3 > value):
            raise ValueError("Invalid triangle: Side lengths do not satisfy the triangle inequality theorem.")
        self._side1 = value

    @property
    def side2(self) -> float:
        return self._side2

    @side2.setter
    def side2(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (self._side1 + value > self._side3 and self._side1 + self._side3 > value and self._side3 + value > self._side1):
            raise ValueError("Invalid triangle: Side lengths do not satisfy the triangle inequality theorem.")
        self._side2 = value

    @property
    def side3(self) -> float:
        return self._side3

    @side3.setter
    def side3(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        if not (self._side1 + self._side2 > value and self._side1 + value > self._side2 and self._side2 + value > self._side1):
            raise ValueError("Invalid triangle: Side lengths do not satisfy the triangle inequality theorem.")
        self._side3 = value

    @property
    def perimeter(self) -> float:
        return self._side1 + self._side2 + self._side3

    @property
    def area(self) -> float:
        s = self.perimeter / 2
        return math.sqrt(s * (s - self._side1) * (s - self._side2) * (s - self._side3))
    
    def __repr__(self) -> str:
        return f"Triangle(side1={self._side1}, side2={self._side2}, side3={self._side3})"

    def __str__(self) -> str:
        return f"Triangle with sides: {self._side1}, {self._side2}, {self._side3}"

# Example Usage
triangle1 = Triangle(3, 4, 5) 
print(triangle1)
print(f"Area: {triangle1.area}")
print(f"Perimeter: {triangle1.perimeter}")

triangle2 = Triangle(5, 5, 5) 
print(triangle2)

#Using Properties
triangle1.side1 = 6
print(triangle1)
print(f"Area: {triangle1.area}")
```

```python
class Trapezoid:

    def __init__(self, base1: float, base2: float, height: float, side1: float, side2: float) -> None:
        if not all(x > 0 for x in [base1, base2, height, side1, side2]):
            raise ValueError("All sides and height must be positive.")
        self._base1 = base1
        self._base2 = base2
        self._height = height
        self._side1 = side1
        self._side2 = side2

    @property
    def base1(self) -> float:
        return self._base1

    @base1.setter
    def base1(self, value: float) -> None:
      if value <= 0:
          raise ValueError("Base must be positive.")
      self._base1 = value

    @property
    def base2(self) -> float:
        return self._base2

    @base2.setter
    def base2(self, value: float) -> None:
      if value <= 0:
          raise ValueError("Base must be positive.")
      self._base2 = value

    @property
    def height(self) -> float:
        return self._height

    @height.setter
    def height(self, value: float) -> None:
      if value <= 0:
          raise ValueError("Height must be positive.")
      self._height = value

    @property
    def side1(self) -> float:
        return self._side1

    @side1.setter
    def side1(self, value: float) -> None:
      """Sets the length of the first non-parallel side."""
      if value <= 0:
          raise ValueError("Side must be positive.")
      self._side1 = value

    @property
    def side2(self) -> float:
        return self._side2

    @side2.setter
    def side2(self, value: float) -> None:
      if value <= 0:
          raise ValueError("Side must be positive.")
      self._side2 = value
      
    @property
    def area(self) -> float:
        return 0.5 * (self._base1 + self._base2) * self._height

    @property
    def perimeter(self) -> float:
        return self._base1 + self._base2 + self._side1 + self._side2

    def __repr__(self) -> str:
        return f"Trapezoid(base1={self._base1}, base2={self._base2}, height={self._height}, side1={self._side1}, side2={self._side2})"

    def __str__(self) -> str:
        return f"Trapezoid with bases: {self._base1}, {self._base2}, height: {self._height}, sides: {self._side1}, {self._side2}"

# Example usage
trap1 = Trapezoid(5, 10, 6, 7, 8)
print(trap1)
print(f"Area: {trap1.area}")
print(f"Perimeter: {trap1.perimeter}")

trap2 = Trapezoid(3, 7, 4, 5, 5)
print(trap2)

#Using the properties
trap1.base1 = 6
print(trap1)
print(f"Area: {trap1.area}")

```

```python
import math

class Parallelogram:

    def __init__(self, base: float, side: float, height: float) -> None:
        if not all(x > 0 for x in [base, side, height]):
            raise ValueError("Base, side, and height must be positive.")
        self._base = base
        self._side = side
        self._height = height

    @property
    def base(self) -> float:
        return self._base

    @base.setter
    def base(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Base must be positive.")
        self._base = value

    @property
    def side(self) -> float:
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side must be positive.")
        self._side = value

    @property
    def height(self) -> float:
        return self._height

    @height.setter
    def height(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Height must be positive.")
        self._height = value

    @property
    def area(self) -> float:
        return self._base * self._height

    @property
    def perimeter(self) -> float:
        return 2 * (self._base + self._side)

    def __repr__(self) -> str:
        return f"Parallelogram(base={self._base}, side={self._side}, height={self._height})"

    def __str__(self) -> str:
        return f"Parallelogram with base: {self._base}, side: {self._side}, height: {self._height}"

# Example Usage
para1 = Parallelogram(5, 7, 4)
print(para1)
print(f"Area: {para1.area}")
print(f"Perimeter: {para1.perimeter}")

para2 = Parallelogram(10, 6, 8)
print(para2)

#Using Properties
para1.base = 8
print(para1)
print(f"Area: {para1.area}")

```

```python
import math

class Rhombus:

    def __init__(self, side: float, angle: float) -> None:
        if side <= 0:
            raise ValueError("Side length must be positive.")
        if not (0 < angle < 180):
            raise ValueError("Angle must be between 0 and 180 degrees (exclusive).")

        self._side = side
        self._angle = angle 
    @property
    def side(self) -> float:
        return self._side

    @side.setter
    def side(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Side length must be positive.")
        self._side = value

    @property
    def angle(self) -> float:
        return self._angle

    @angle.setter
    def angle(self, value: float) -> None:
        if not (0 < value < 180):
            raise ValueError("Angle must be between 0 and 180 degrees (exclusive).")
        self._angle = value

    @property
    def area(self) -> float:
        angle_rad = math.radians(self._angle) # Convert to radians for math.sin
        return self._side**2 * math.sin(angle_rad)

    @property
    def perimeter(self) -> float:
        return 4 * self._side

    @property
    def diagonal1(self) -> float:
        angle_rad = math.radians(self._angle)
        return 2 * self._side * math.cos(angle_rad / 2)

    @property
    def diagonal2(self) -> float:
        angle_rad = math.radians(self._angle)
        return 2 * self._side * math.sin(angle_rad / 2)


    def __repr__(self) -> str:
        return f"Rhombus(side={self._side}, angle={self._angle})"

    def __str__(self) -> str:
        return f"Rhombus with side: {self._side}, angle: {self._angle} degrees"

# Example Usage
rhombus1 = Rhombus(5, 60)
print(rhombus1)
print(f"Area: {rhombus1.area}")
print(f"Perimeter: {rhombus1.perimeter}")
print(f"Diagonal 1: {rhombus1.diagonal1}")
print(f"Diagonal 2: {rhombus1.diagonal2}")

rhombus2 = Rhombus(7, 90)
print(rhombus2)

#Using Properties
rhombus1.side = 8
print(rhombus1)
print(f"Area: {rhombus1.area}")

rhombus1.angle = 30
print(rhombus1)
print(f"Area: {rhombus1.area}")
```

```python
class Resistor:


    def __init__(self, resistance: float) -> None:
        if resistance <= 0:
            raise ValueError("Resistance must be positive.")
        self._resistance = resistance

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Resistance must be positive.")
        self._resistance = value

    def voltage_drop(self, current: float) -> float:
        return self._resistance * current

    def power_dissipated(self, current: float) -> float:
      return (current**2) * self._resistance

    def __repr__(self) -> str:
        return f"Resistor(resistance={self._resistance})"

    def __str__(self) -> str:
        return f"Resistor: {self._resistance} ohms"

# Example usage:
r1 = Resistor(100)  # A 100-ohm resistor
print(r1)
print(f"Resistance: {r1.resistance} ohms")

current = 0.1  # 0.1 amperes
voltage_drop = r1.voltage_drop(current)
print(f"Voltage drop across r1 at {current}A: {voltage_drop} volts")

power = r1.power_dissipated(current)
print(f"Power dissipated by r1 at {current}A: {power} watts")

r2 = Resistor(220)
print(r2)

#Using Properties
r1.resistance = 150
print(r1)
print(f"Voltage drop across r1 at {current}A: {r1.voltage_drop(current)} volts")
```

```python
import cmath  
import math

class RL:

    def __init__(self, resistance: float, inductance: float, angular_frequency: float = 1) -> None:
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if inductance < 0:
            raise ValueError("Inductance must be non-negative.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Inductance must be non-negative.")
        self._inductance = value
    
    @property
    def angular_frequency(self) -> float:
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def impedance(self) -> complex:
        j = complex(0, 1)  # Imaginary unit
        return self._resistance + j * (self._angular_frequency * self._inductance)

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    @property
    def power_factor(self) -> float:
        return math.cos(self.phase)

    def current(self, voltage: complex) -> complex:
        if self.impedance == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance

    def __repr__(self) -> str:
        return f"RL(R={self._resistance}, L={self._inductance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        return f"RL Circuit: R={self._resistance} Ω, L={self._inductance} H, ω={self._angular_frequency} rad/s"

# Example usage:
rl_circuit = RL(10, 0.1, 50) 
print(rl_circuit)
print(f"Impedance: {rl_circuit.impedance}")
print(f"Phase Angle: {rl_circuit.phase} radians")
print(f"Power Factor: {rl_circuit.power_factor}")

voltage = complex(10, 0)  
current = rl_circuit.current(voltage)
print(f"Current: {current}")

rl_circuit2 = RL(20, 0.2)

print(rl_circuit2)

#Using Properties
rl_circuit.resistance = 15
print(rl_circuit)
print(f"Impedance: {rl_circuit.impedance}")

```

```python
import cmath  
import math

class RC:

    def __init__(self, resistance: float, capacitance: float, angular_frequency: float = 1) -> None:
        if resistance < 0:
            raise ValueError("Resistance must be non-negative.")
        if capacitance < 0:
            raise ValueError("Capacitance must be non-negative.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._capacitance = capacitance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Resistance must be non-negative.")
        self._resistance = value

    @property
    def capacitance(self) -> float:
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        if value < 0:
            raise ValueError("Capacitance must be non-negative.")
        self._capacitance = value

    @property
    def angular_frequency(self) -> float:
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value

    @property
    def impedance(self) -> complex:
        j = complex(0, 1)  # Imaginary unit
        omega = self.angular_frequency
        if self.capacitance == 0:
            return complex(self.resistance)
        return self.resistance - j / (omega * self.capacitance)

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    @property
    def power_factor(self) -> float:
        return math.cos(self.phase)

    def current(self, voltage: complex) -> complex:
        if self.impedance == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / self.impedance

    def __repr__(self) -> str:
        return f"RC(R={self._resistance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        return f"RC Circuit: R={self._resistance} Ω, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

# Example usage:
rc_circuit = RC(1000, 1e-6, 1000) 
print(rc_circuit)
print(f"Impedance: {rc_circuit.impedance}")
print(f"Phase Angle: {rc_circuit.phase} radians")
print(f"Power Factor: {rc_circuit.power_factor}")

voltage = complex(5, 0)  
current = rc_circuit.current(voltage)
print(f"Current: {current}")

rc_circuit2 = RC(2000, 2e-6)

print(rc_circuit2)

#Using Properties
rc_circuit.resistance = 1500
print(rc_circuit)
print(f"Impedance: {rc_circuit.impedance}")

```

```python
import cmath
import math

class RLCParallel:


    def __init__(self, resistance: float, inductance: float, capacitance: float, angular_frequency: float = 1) -> None:
        if resistance <= 0:
            raise ValueError("Resistance must be positive.")
        if inductance <= 0:
            raise ValueError("Inductance must be positive.")
        if capacitance <= 0:
            raise ValueError("Capacitance must be positive.")
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance
        self._angular_frequency = angular_frequency

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Resistance must be positive.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Inductance must be positive.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Capacitance must be positive.")
        self._capacitance = value

    @property
    def angular_frequency(self) -> float:
        return self._angular_frequency

    @angular_frequency.setter
    def angular_frequency(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Angular frequency must be positive.")
        self._angular_frequency = value
        
    @property
    def admittance(self) -> complex:
        j = complex(0, 1)
        omega = self._angular_frequency
        if self._inductance == 0:
            bl = 0
        else:
            bl = 1/(omega * self._inductance)
        if self._capacitance == 0:
            bc = 0
        else:
            bc = omega * self._capacitance
            
        return 1 / self._resistance + j * (bc - bl)

    @property
    def impedance(self) -> complex:
        if self.admittance == 0:
            return complex(0)
        return 1 / self.admittance

    @property
    def phase(self) -> float:
        return cmath.phase(self.impedance)

    @property
    def power_factor(self) -> float:
        return math.cos(self.phase)

    def current(self, voltage: complex) -> complex:
        if self.admittance == 0:
            raise ZeroDivisionError("Admittance cannot be zero.")
        return voltage * self.admittance

    def __repr__(self) -> str:
        return f"RLCParallel(R={self._resistance}, L={self._inductance}, C={self._capacitance}, ω={self._angular_frequency})"

    def __str__(self) -> str:
        return f"Parallel RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F, ω={self._angular_frequency} rad/s"

# Example usage:
rlc_parallel = RLCParallel(10, 0.1, 1e-6, 1000) 
print(rlc_parallel)
print(f"Admittance: {rlc_parallel.admittance}")
print(f"Impedance: {rlc_parallel.impedance}")
print(f"Phase Angle: {rlc_parallel.phase} radians")
print(f"Power Factor: {rlc_parallel.power_factor}")

voltage = complex(10, 0)  # 10 V at 0 degrees phase
current = rlc_parallel.current(voltage)
print(f"Total Current: {current}")

rlc_parallel2 = RLCParallel(20, 0.2, 2e-6)

print(rlc_parallel2)

#Using Properties
rlc_parallel.resistance = 15
print(rlc_parallel)
print(f"Admittance: {rlc_parallel.admittance}")

```

```python
import cmath
import math

class RLCSeries:

    def __init__(self, resistance: float, inductance: float, capacitance: float) -> None:
        if resistance <= 0:
            raise ValueError("Resistance must be positive.")
        if inductance <= 0:
            raise ValueError("Inductance must be positive.")
        if capacitance <= 0:
            raise ValueError("Capacitance must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Resistance must be positive.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Inductance must be positive.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Capacitance must be positive.")
        self._capacitance = value

    @property
    def resonant_frequency(self) -> float:
        if self._inductance == 0 or self._capacitance == 0:
            return float('inf') #If either L or C is zero, the resonant frequency is infinite
        return 1 / math.sqrt(self._inductance * self._capacitance)

    @property
    def resonant_frequency_hz(self) -> float:
        return self.resonant_frequency / (2 * math.pi)

    def impedance(self, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        j = complex(0, 1)
        return self._resistance + j * (angular_frequency * self._inductance - 1 / (angular_frequency * self._capacitance))

    def current(self, voltage: complex, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        z = self.impedance(angular_frequency)
        if z == 0:
            raise ZeroDivisionError("Impedance cannot be zero.")
        return voltage / z

    def __repr__(self) -> str:
        return f"RLCSeries(R={self._resistance}, L={self._inductance}, C={self._capacitance})"

    def __str__(self) -> str:
        return f"Series RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F"


# Example Usage
rlc = RLCSeries(10, 0.1, 1e-6)
print(rlc)

print(f"Resonant Frequency (rad/s): {rlc.resonant_frequency}")
print(f"Resonant Frequency (Hz): {rlc.resonant_frequency_hz}")

omega = rlc.resonant_frequency #at resonance impedance is minimum and equal to Resistance
voltage = complex(10, 0)
current = rlc.current(voltage, omega)
print(f"Current at resonance: {current}")

omega2 = 10000 #off resonance
current2 = rlc.current(voltage, omega2)
print(f"Current off resonance: {current2}")

#Using Properties
rlc.resistance = 20
print(rlc)
```

```python
import cmath
import math

class RLCParallel:

    def __init__(self, resistance: float, inductance: float, capacitance: float) -> None:
        if resistance <= 0:
            raise ValueError("Resistance must be positive.")
        if inductance <= 0:
            raise ValueError("Inductance must be positive.")
        if capacitance <= 0:
            raise ValueError("Capacitance must be positive.")

        self._resistance = resistance
        self._inductance = inductance
        self._capacitance = capacitance

    @property
    def resistance(self) -> float:
        return self._resistance

    @resistance.setter
    def resistance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Resistance must be positive.")
        self._resistance = value

    @property
    def inductance(self) -> float:
        return self._inductance

    @inductance.setter
    def inductance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Inductance must be positive.")
        self._inductance = value

    @property
    def capacitance(self) -> float:
        return self._capacitance

    @capacitance.setter
    def capacitance(self, value: float) -> None:
        if value <= 0:
            raise ValueError("Capacitance must be positive.")
        self._capacitance = value

    @property
    def resonant_frequency(self) -> float:
        if self._inductance == 0 or self._capacitance == 0:
            return float('inf') #If either L or C is zero, the resonant frequency is infinite
        return 1 / math.sqrt(self._inductance * self._capacitance)

    @property
    def resonant_frequency_hz(self) -> float:
        return self.resonant_frequency / (2 * math.pi)

    def admittance(self, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        j = complex(0, 1)
        if self._inductance == 0:
            bl = 0
        else:
            bl = 1/(angular_frequency * self._inductance)
        if self._capacitance == 0:
            bc = 0
        else:
            bc = angular_frequency * self._capacitance
        return 1 / self._resistance + j * (bc - bl)

    def impedance(self, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        y = self.admittance(angular_frequency)
        if y == 0:
            return complex(0) #Return 0 impedance if admittance is zero
        return 1 / y

    def current(self, voltage: complex, angular_frequency: float) -> complex:
        if angular_frequency <= 0:
            raise ValueError("Angular frequency must be positive.")
        y = self.admittance(angular_frequency)
        return voltage * y

    def __repr__(self) -> str:
        return f"RLCParallel(R={self._resistance}, L={self._inductance}, C={self._capacitance})"

    def __str__(self) -> str:
        return f"Parallel RLC Circuit: R={self._resistance} Ω, L={self._inductance} H, C={self._capacitance} F"


# Example usage:
rlc_parallel = RLCParallel(10, 0.1, 1e-6)  # 10 ohms, 0.1 H, 1 µF
print(rlc_parallel)

print(f"Resonant Frequency (rad/s): {rlc_parallel.resonant_frequency}")
print(f"Resonant Frequency (Hz): {rlc_parallel.resonant_frequency_hz}")

omega = rlc_parallel.resonant_frequency #at resonance admittance is minimum and impedance is maximum
voltage = complex(10, 0)
current = rlc_parallel.current(voltage, omega)
print(f"Current at resonance: {current}")

omega2 = 10000 #off resonance
current2 = rlc_parallel.current(voltage, omega2)
print(f"Current off resonance: {current2}")

#Using Properties
rlc_parallel.resistance = 15
print(rlc_parallel)
```
